# Split images

```python
# given 1D image representation return the image reshaped into a grid
def imaged_grid(img , row , col ):
    """
    Input:
        img: 1D Image vector
        row: number of rows in the grid
        col: number of columns in the grid
    Output:
        reshaped image (2D)
    """
    x , y = img.shape

    assert x % row == 0, "{} rows is not evenly divisble by {}".format(x, row)
    assert y % col == 0, "{} cols is not evenly divisble by {}".format(y, col)
    
    return (img.reshape(x//row, row, -1,col)
            .swapaxes(1,2)
            .reshape(-1, row, col)) 
```

# Feature extraction (using centroid)

```python
def get_centroid(img):
    """
    Input: 
        img: (2D) image
    
    Output: 
        feature vector of the img (np.array)
    """
    
    feature_vector = []
 
    for grid in imaged_grid(img ,2, 2) :
        
        Xc = 0 
        Yc = 0 
        sum = 0
    
        for index, x in np.ndenumerate(grid):
          sum+= x 
          Xc += x * index[0]
          Yc += x * index[1]
        
        if sum != 0 :
            feature_vector.append( Xc/ sum )
            feature_vector.append(Yc/ sum )
        else :
             feature_vector.append(0)
             feature_vector.append(0)
        
    
    return np.array(feature_vector)
```

# fit and predict

```python
def KNN(train_features, test_features, train_labels):
     """
     Inputs: 
         train_features: feature vector of images in training data
         test_features: feature vector of images in test data
         train_labels: labels of images in trainig data
         
     Output:
         return the prediction of labels of test images 
     """
    knn = KNeighborsClassifier(4, metric='euclidean')
    knn.fit(train_features, train_labels)  # fit train data
    prediction = knn.predict(test_features)  # test data
    return prediction
```

```python
Knn_prediction = KNN(train_features, test_features , y_train )
```

# Accuracy

```python
print("Accuracy Score =", accuracy_score(y_test, Knn_prediction) * 100, "%")
```
